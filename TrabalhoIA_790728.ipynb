```python
#Importar o algoritmo de associação
from efficient_apriori import apriori
```

```python
#Le o arquvio em formato csv
import pandas as pd
df = pd.read_csv('house-votes-84.data',header = None)
df.head()
```

Esses 3 laços irão modificar o dataset para se adequar ao algoritmo.

O primeiro laço, troca o "?" que representa um valor inexistente por "nan", para se adequar à notação comum

O segundo laço adiciona o número da coluna ao valor y, de cada coluna, para poder identificar de qual coluna é o valor, para facilitar no entendimento das regras de associação

O terceiro laço adiciona o número da coluna ao valor n, de cada coluna, para poder identificar de qual coluna é o valor, para facilitar no entendimento das regras de associação

```python
for i in range(1, df.shape[1]):
  df[i] = df[i].replace(['?'],'nan')
```

```python
for i in range(1, df.shape[1]):
  df[i] = df[i].replace(['y'],f'y{i}')
```

```python
for i in range(1, df.shape[1]):
  df[i] = df[i].replace(['n'],f'n{i}')
```

Coluna da classe é retirada

```python
df = df.drop([0],axis=1)
```

Cabeçalho do dataframe é exibido, mostrando as alterações realizadas

```python
#Transformar os dados para listas de listas
t_list = []
for i in range(0, 100):
  l_in = []
  for j in range(0,9):
    s = str(df.values[i,j])
    if (s != 'nan'):
      l_in.append(s)
  t_list.append(l_in)
```

Pega os itemsets e as regras

```python
itemsets, rules = apriori(t_list, min_support=0.05, min_confidence = 0.3)
```
